## 6. Notre variable target (Y) est le RainTomorrow, Récuprer X et Y à partir du jeu de données 

```python
df.head(2)
```

```python
y = df['RainTomorrow']
X = df.drop(['RainTomorrow'], axis=1)

print(X.shape, y.shape)
```

## 7. Fractionnement le jeu de données en jeu d'entraînement et jeu de test (20% pour le test)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
print("train set :", X_train.shape) 
print("test set :", X_test.shape)
```

## 8. Checker les valeurs manquantes des variables quantitatives du jeu d'entraînement et de test

```python
# Récupérer tout d'abord les variables quantitatives (numerical)
numerical = []
for col in X.columns :
    if X[col].dtypes != 'O':
        numerical.append(col)


# check missing values in numerical variables in X_train
# check missing values in numerical variables in X_test
print(">>> Valeurs manquantes :")
print("Ttrain set :")
print(X_train[numerical].isnull().sum())
print("=======================\nTest set :")
print(X_test[numerical].isnull().sum())
```

## 9. En utilisant SimpleImputer de Scikit-Learn remplacer les valeurs manquantes dans les variables numérique (entraînement et test) par la moyenne

```python
from sklearn.impute import SimpleImputer

Train_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_train[numerical] = Train_imputer.fit_transform(X_train[numerical]) 
Test_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_test[numerical] = Test_imputer.fit_transform(X_test[numerical]) 
```

## 10. Vérifier que les valeurs manquantes n'existent plus dans les variables quantitatives (entraînement et test)

```python
# check missing values in numerical variables in X_train
# check missing values in numerical variables in X_test
print(">>> Valeurs manquantes :")
print("Ttrain set :")
print(X_train[numerical].isnull().sum())
print("=======================\nTest set :")
print(X_test[numerical].isnull().sum())
```

## 11. En utilisant la fonction fillna() et mode() remplacer les valeurs manquantes dans les variables qualitatives par la valeur la plus fréquante pour chaque variable.

```python
# Récuperer tout d'abord les variables qualitatives
categorial = []
for col in X.columns :
    if X[col].dtypes == 'O':
        categorial.append(col)
```

```python
print(">>> Valeurs manquantes :")
print("Ttrain set :")
print(X_train[categorial].isnull().sum())
print("=======================\nTest set :")
print(X_test[categorial].isnull().sum())
```

```python
# visualisation des valeurs les plus fréquentes par colonne (Mode)
print(">>> Modes :")
print("Ttrain set :")
print(X_train[categorial].mode())
print("============================================================\nTest set :")
print(X_test[categorial].mode())
```

```python
for dataset in [X_train, X_test]:
    for col in categorial :
        dataset[col].fillna(value=X_train[col].mode().iloc[0], inplace=True)
```

## 13. Encoder la variable qualitative 'RainToday' en utilisant le module category_encoders et la fonction BinaryEncoder(cols=['RainToday']) pour le jeu d'entraînement et de test.

```python
# Installation: Dans Anaconda Prompt tapez la commande: pip install category_encoders
```

```python
import category_encoders as ce

binary_encoder = ce.BinaryEncoder(cols=['RainToday'])

X_train = binary_encoder.fit_transform(X_train)

X_test = binary_encoder.transform(X_test)
```

```python
print(X_train.shape)
X_train.head(5)
```

## 14. Recréer un nouveau jeu d'entrainement en concatenant les variables numériques du jeu d'entrainement précedent avec les deux colonnes RainToday_0', 'RainToday_1'qui vont se créer à l'aide de l'étape précedente, plus les variables qualitatives transformées à des variables muettes avec la fonction get_dummies() de pandas 

```python
#print(pd.get_dummies(X_train.Location).shape)
pd.get_dummies(X_train.Location)
```

```python
# Variables catégorielles :
# Location WindGustDir WindDir9am  WindDir3pm  RainToday  

# RainToday   > Binary encoder  

# Pour get_dummies, il reste les colonnes Location WindGustDir WindDir9am WindDir3pm à utiliser

xtmp = pd.concat([pd.get_dummies(X_train.Location), pd.get_dummies(X_train.WindGustDir), pd.get_dummies(X_train.WindDir9am), pd.get_dummies(X_train.WindDir3pm)], axis=1)

print(xtmp.shape)
```

```python
print(X_train[numerical].shape)
print(X_train[['RainToday_0', 'RainToday_1']].shape)
print(xtmp.shape)
```

```python
X_train = pd.concat([X_train[numerical], X_train[['RainToday_0', 'RainToday_1']], xtmp], axis=1)

X_train.shape
```

## 15. Refaire l'étape précedente pour le jeu de test

```python
X_test = pd.concat([X_test[numerical], X_test[['RainToday_0', 'RainToday_1']], pd.get_dummies(X_test.Location), pd.get_dummies(X_test.WindGustDir), pd.get_dummies(X_test.WindDir9am), pd.get_dummies(X_test.WindDir3pm)], axis=1)

X_test.shape
```

```python
# train a logistic regression model on the training set
# instantiate the model
# fit the model

from sklearn.linear_model import LogisticRegression

regressor = LogisticRegression(solver='liblinear')
regressor.fit(X_train, y_train)
```

##  19. Afficher la matrice de confusion, TP, TN, FP, FN sur le jeu du test en utilisant la fonction confusion_matrix


```python
from sklearn.metrics import confusion_matrix
```

```python
cmat = confusion_matrix(y_test, y_pred)
print(cmat)

print("TP =", cmat[0][0])
print("TN =", cmat[1][1])
print("FP =", cmat[0][1])
print("FN =", cmat[1][0])
```

##  20. Calculer les métriques de classification (accuracy, classification error, precision, recall, specificity) en utilisant seulement les valeurs de TP, TN, FP, FN.

```python
TP = cmat[0][0]
TN = cmat[1][1]
FP = cmat[0][1]
FN = cmat[1][0]
```

```python
# accuracy
accuracy = round((TP+TN)/(TP+FP+FN+TN),4)

# classification error
error_rate = round((FP + FN) / (TP + TN + FP + FN),4)

# precision
precision = round(TP/(TP+FP),4)

# recall
recall = round(TP/(TP+FN),4)

# specificity
specificity = round(TN/(TN+FP),4)

# F1-score
f1_score = round(2*(recall * precision) / (recall + precision),4)

print("Classification accuracy :" ,accuracy, "\nClassification error :", error_rate, "\nPrecision :", precision, "\nRecall or Sensitivity :", recall, "\nSpecificity :", specificity)
```
