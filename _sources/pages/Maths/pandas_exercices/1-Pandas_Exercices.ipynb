## Étape 2. Attribuez-le à une variable appelée drinks.

```python
drinks=pd.read_csv("../pandas_exercices/drinks.csv", sep=",")
drinks
```

## Étape 3. Quel continent boit plus de l'alcool en moyenne?

```python
tmp = drinks.groupby(drinks["continent"])
tmp = tmp.sum()
tmp = tmp.sort_values(by = ["total_litres_of_pure_alcohol"], ascending = False)
tmp.head(1)
```

```python
drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean().idxmax()
```

## Étape 4. Pour chaque continent, afficher les statistiques de la consommation de l'alcool.

```python
tmp2 = drinks.total_litres_of_pure_alcohol.groupby(drinks["continent"])
tmp2.describe()
```

## Étape 5. Afficher la consommation moyenne d'alcool par continent pour chaque colonne

```python
drinks.groupby('continent').mean()
```

## Étape 6. Afficher la consommation médiane d'alcool par continent pour chaque colonne

```python
drinks.groupby('continent').median()
```

## Étape 7. Afficher les valeurs moyenne, minimale et maximale de la consommation d'alcool.

```python
drinks.describe().loc[['mean','max','min']]
```
