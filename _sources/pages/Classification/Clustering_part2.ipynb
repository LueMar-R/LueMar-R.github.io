```python
index = np.random.permutation(60000)[:10000]
print(index)
```

### 8. Afin de réduire la dimensionnalité des données, appliquez la méthode PCA sur les données (X) en choisissant un nombre de composants de 2. 
Pensez à utiliser la fonction
fit_transform (et non pas fit) pour stocker les données réduites dans une variable.

```python
model = PCA(n_components=2)
```

```python
reduced_X = model.fit_transform(X)
```

### 9. Affichez les données réduites (les données après l'application de PCA) en suivant les étapes suivantes :
